## Парсинг любимых анекдотов с anecdot.ru

для начала рассмотрим лучший анекдоты за месяц

```python
import requests
req = requests.get('https://www.anekdot.ru/release/anekdot/month/')
print req.text
```

Чтобы нас пустили на сайт нужно использовать user agent

```python
headers = {'User-agent': 'Mozilla/5.0'}
req = requests.get('https://www.anekdot.ru/release/anekdot/month/', headers=headers)
```

```python
import bs4
parser = bs4.BeautifulSoup(req.text, 'lxml')
print parser
```

ищем нужный класс div, содержащий уморительный анекдот

```python
print parser.find('div', attrs={'class':'text'}).text
```

также можно найти объект, содержайщий рейтинг анекдота: в массиве будет итоговый рейтинг, полное количество голосов, кол-во голосов за и против

```python
print parser.find('div', attrs={'class':'rates'})['data-r'].split(';')
```

```python
anecs = parser.findAll('div', attrs={'class':'text'})
votes = parser.findAll('div', attrs={'class':'rates'})
```

```python
for a,v in zip(anecs, votes):
    print a.text
    print "Рейтинг: " + v['data-r'].split(";")[0]
    print "Всего голосов: " + v['data-r'].split(";")[1]
    print "Голосов 'за': " + v['data-r'].split(";")[2]
    print "Голосов 'против': " + v['data-r'].split(";")[3]
    print "\n------\n"
```

для создания датасета возьмем страницу со случайными анекдотами: так там будут анекдоты и с высоким и с низким рейтингом

```python
req = requests.get('https://www.anekdot.ru/random/anekdot/', headers=headers)
parser = bs4.BeautifulSoup(req.text, 'lxml')
```

```python
anecs = parser.findAll('div', attrs={'class':'text'})
votes = parser.findAll('div', attrs={'class':'rates'})

for a,v in zip(anecs, votes):
    print a.text
    print "Рейтинг: " + v['data-r'].split(";")[0]
    print "Всего голосов: " + v['data-r'].split(";")[1]
    print "Голосов 'за': " + v['data-r'].split(";")[2]
    print "Голосов 'против': " + v['data-r'].split(";")[3]
    print "\n------\n"
```

создаем датасет со все информацией

```python
import pandas as pd
df = pd.DataFrame(columns=['text', 'vote', 'total', '+', '-'])
df
```

```python
text = []
for a in anecs:
    text.append(a.text)
df.text = text  

for v in range(0, len(votes)):
    df.loc[v][1:5] = votes[v]['data-r'].split(";")
    
df.head()
```
